```python
plt.figure(dpi=dpi_plot)
plt.title("Sumowanie od początku")
plt.ylabel("Błąd względny")
plt.xlabel("Kąt [$^\circ$]")
plt.yscale('symlog', linthreshy=1e-16)
plt.tight_layout()
plt.xticks(range(0,181,10),rotation=70)
plt.plot(bw[['x']],bw[['mysin']], color = 'blue',linestyle='-')
```

```python
plt.figure(dpi=dpi_plot)
plt.title("Sumowanie od końca")
plt.ylabel("Błąd względny")
plt.xlabel("Kąt [$^\circ$]")
plt.yscale('symlog', linthreshy=1e-16)
plt.tight_layout()
plt.xticks(range(0,181,10),rotation=70)
plt.plot(bw[['x']],bw[['mysinrev']], color = 'green',linestyle='-')
```

```python
plt.figure(dpi=dpi_plot)
plt.title("Obliczanie wyrazu na podstawie poprzednika od początku")
plt.ylabel("Błąd względny")
plt.xlabel("Kąt [$^\circ$]")
plt.yscale('symlog', linthreshy=1e-16)
plt.tight_layout()
plt.xticks(range(0,181,10),rotation=70)
plt.plot(bw[['x']],bw[['mysinterms']], color = 'red',linestyle='-')

```

```python
plt.figure(dpi=dpi_plot)
plt.title("Obliczanie wyrazu na podstawie poprzednika od końca")
plt.ylabel("Błąd względny")
plt.xlabel("Kąt [$^\circ$]")
plt.yscale('symlog', linthreshy=1e-16)
plt.tight_layout()
plt.xticks(range(0,181,10),rotation=70)
plt.plot(bw[['x']],bw[['mysintermsrev']], color = 'red',linestyle='-')
```

```python
plt.figure(dpi=dpi_plot)
plt.title("Pórównianie błedów najdokładniejszych metod")
plt.ylabel("Błąd względny")
plt.xlabel("Kąt [$^\circ$]")
plt.yscale('symlog', linthreshy=1e-16)
# plt.tight_layout()
plt.xticks(range(0,181,10),rotation=70)
plt.plot(bw[['x']],bw[['mysinterms']], color = 'green',linestyle='-', label = 'Szereg potęgowy od końca')
plt.plot(bw[['x']],bw[['mysintermsrev']], color = 'orange',linestyle='-',  label = 'Na podstawie poprzednika od końca')
plt.xlim([90,180])

plt.legend()
```

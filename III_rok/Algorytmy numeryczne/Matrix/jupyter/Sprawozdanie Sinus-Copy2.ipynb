```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

dpi_plot = 200

matplotlib.style.use('ggplot')

result = pd.read_csv('../bledy.csv', delimiter=',')

# er_sin_szereg_rev = pd.read_csv('../szereg_od_tylu.csv', delimiter=',')
# er_sin_term = pd.read_csv('../szereg_term.csv', delimiter=',')
```

```python
plt.figure(dpi=dpi_plot)
plt.title("$A \cdot X$")
plt.ylabel("Bd wzgldny")
plt.xlabel("Rozmiar macierzy")
plt.yscale('symlog', linthreshy=0.0000000000000001)
plt.plot(result[['n']],result[['fr1B']], color = 'red', linestyle="-", label='Uamek')
plt.plot(result[['n']],result[['fl1B']], color = 'blue', linestyle="-", label='Float')
plt.plot(result[['n']],result[['d1B']], color = 'green', linestyle="-", label='Double')


plt.legend()
# plt.savefig('sample.pdf')
```

```python
plt.figure(dpi=dpi_plot)
plt.title("$A \cdot X$")
plt.ylabel("Czas wykonywania $[s]$")
plt.xlabel("Rozmiar macierzy")
plt.yscale('symlog', linthreshy=0.0000000000000001)
plt.plot(result[['n']],result[['fr1T']], color = 'red', linestyle="-", label='Uamek')
plt.plot(result[['n']],result[['fl1T']], color = 'blue', linestyle="-", label='Float')
plt.plot(result[['n']],result[['d1T']], color = 'green', linestyle="-", label='Double')
plt.plot(result[['n']],result[['e1T']], color = 'yellow', linestyle="-", label='Eigen')



plt.legend()
# plt.savefig('sample.pdf')
```
